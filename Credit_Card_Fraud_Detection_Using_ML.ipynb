importing dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data=pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
#print("Missing values:\n", missing_values)
```

```python
data = credit_card_data.fillna(credit_card_data.mode().iloc[0])
print("Cleaned dataset:\n", data.tail())
```

```python
# distribution of legit trasactions and fraudulent transactions
data['Class'].value_counts()
```

# This datset if highly unbalanced
# 0 --> Normal Transaction
# 1 --> Fraudulent transaction

```python
#Separating data for analysis
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both  transactions
credit_card_data.groupby('Class').mean()
```

```python
#Under-Sampling
# Build a sample dataset containg similar distribution of normal trasaction and Fradulent Trasactions
# Number of Fradulent Trasactions --> 360
```

```python
legit_sample = legit.sample(n=360)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
# axis = 0 --> row
# axis = 1 --> column
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
#Important step --> As we get know if we got a good sample or bad
```

```python
#Splitting the data into  features and Targets
```

```python
X = new_dataset.drop(columns='Class', axis = 1)
Y = new_dataset['Class']
```

```python
X
```

```python
Y
```

```python
#Split the data into Training data and Testing data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
#Features present in X
#Labels present in Y
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Model training --> Logistic Regression for binary classfication problems
```

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression with Training Data
model.fit(X_train, Y_train)
```

```python
# Model Evaluation
#Accuracy score
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
#Accuracy score on Training data:
print('Accuracy on Training data:', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
#Accuracy score on Test data:
print('Accuracy on Test data:', test_data_accuracy)
```
